```python
import pandas as pd
from matplotlib import pyplot as plt
import os
import numpy as np
```

```python
submissions = sorted(os.listdir("submissions"))
sub_shapes = []
```

## Visualizations

```python
fig, ax = plt.subplots(len(submissions), 1, figsize=(20, 20))

for i, axis in enumerate(ax.flatten()):
    sub = pd.read_csv("submissions/" + submissions[i])
    sub_shapes.append(sub.shape)
    axis.set_title(submissions[i])
    axis.hist(sub['target'], bins=1000)
    axis.hist(sub['target'], bins=1000)
```

## Check submissions shapes

```python
assert (np.allclose(sub_shapes, sub_shapes))
print("OK!")
```
