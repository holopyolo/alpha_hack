```python
import pandas as pd
from matplotlib import pyplot as plt
import os
```

```python
submissions = sorted(os.listdir("submissions"))
```

```python
fig, ax = plt.subplots(len(submissions), 1, figsize=(12, 12))

for i, axis in enumerate(ax.flatten()):
    sub = pd.read_csv("submissions/" + submissions[i])
    axis.set_title(submissions[i])
    axis.hist(sub['target'], bins=1000)
    axis.hist(sub['target'], bins=1000)
```

```python
fig, ax = plt.subplots(len(submissions), 1, figsize=(12, 12))

for i, axis in enumerate(ax.flatten()):
    sub = pd.read_csv("submissions/" + submissions[i])
    axis.set_title(submissions[i])
    axis.hist(sub['target'][sub['target'] > 0.8], bins=1000)
    axis.hist(sub['target'][sub['target'] > 0.8], bins=1000)
```
